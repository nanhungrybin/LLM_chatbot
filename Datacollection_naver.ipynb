```python
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException

url_list = []
title_list = []

# 검색어 입력
search_query = "사출성형"  # 여기에 원하는 검색어를 입력하세요
search_box = driver.find_element(By.ID, "query")
search_box.send_keys(search_query)
search_box.send_keys(Keys.RETURN)

# 페이지가 로딩될 때까지 잠시 대기
time.sleep(3)

# view 클릭
# CSS 선택자를 사용하여 "블로그" 링크를 찾아 클릭
blog_link_css = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, "a.tab[href*='tab.blog.all']"))
)
blog_link_css.click()


# 스크롤 밑으로 내려주는 함수 
def scroll_to_bottom(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")
    
    while True:
        # 끝까지 스크롤 다운
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        
        try:
            # 페이지 로드를 기다림 (대기 시간을 10초로 늘림)
            WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.TAG_NAME, "footer")))
        except TimeoutException:
            # footer가 지정된 시간 내에 나타나지 않는 경우, 스크롤이 계속되도록 예외 처리
            print("Footer not visible within the timeout period, but continuing the scrolling.")
        
        # 새로운 높이 계산
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        
        last_height = new_height
        
# 사용
scroll_to_bottom(driver)
    
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 블로그 글 URL과 제목을 수집할 리스트를 초기화
url_list = []
title_list = []
content_list = []

# 블로그 글에 대한 수정된 CSS 선택자
articles_selector = ".title_link"
contents_selector = ".dsc_link"

# WebDriverWait을 사용하여 해당 요소들이 로드될 때까지 기다림
try:
    articles_raw = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, articles_selector))
    )
    cont_raw = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, contents_selector))
    )
    
    # 각 요소(링크)에서 URL과 제목을 추출
    for article in articles_raw:
        url = article.get_attribute("href")  # href 속성을 통해 URL 추출
        title = article.text  # 링크 텍스트를 통해 제목 추출
        
        # 추출된 URL과 제목을 리스트에 추가
        if url:  # href 속성이 존재할 경우에만 리스트에 추가
            url_list.append(url)
            title_list.append(title)
            
    for content in cont_raw:
        content = content.text
        content_list.append(content)
    
    # 수집된 데이터를 출력
    for title, url, content in zip(title_list, url_list, content_list):
        print(f"Title: {title}, URL: {url}, content : {content}")
    
    print(" ")
    print(f"URL 개수: {len(url_list)}")
    print(f'제목 개수: {len(title_list)}')
    
    # DataFrame으로 변환하여 저장
    df = pd.DataFrame({'URL': url_list, 'Title': title_list, 'content' : content_list})
    print(df)
    df.to_csv(f"blog_links.csv", index=False)

except Exception as e:
    print(f"오류 발생: {e}")
```

```python
#<a target="_blank" href="https://terms.naver.com/entry.naver?docId=412196&amp;cid=42327&amp;categoryId=42327" class="lnk_tit" onclick="return goOtherCR(this,'a=kdc.title&amp;r=1&amp;i=0810AA45_003008cbfcfc&amp;u='+urlencode(this.href));"><mark>사출 성형</mark></a>

import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

knowledge_base_link_css = "a.tab[href*='tab.kdic.all']"

# WebDriverWait을 사용하여 "지식백과" 링크가 클릭 가능할 때까지 최대 10초간 기다림
knowledge_base_link = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, knowledge_base_link_css))
)

# "지식백과" 링크 클릭
knowledge_base_link.click()

# 스크롤로 끝까지 내리기
scroll_to_bottom(driver)

# 지식백과 항목 URL과 제목을 수집할 리스트를 초기화
url_list = []
title_list = []

# 지식백과 항목에 대한 수정된 CSS 선택자
articles_selector = "a.lnk_tit"  # 주어진 HTML 요소를 기반으로 수정

# WebDriverWait을 사용하여 해당 요소들이 로드될 때까지 기다림
try:
    articles_raw = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, articles_selector))
    )
    
    # 각 요소(링크)에서 URL과 제목을 추출
    for article in articles_raw:
        url = article.get_attribute("href")  # href 속성을 통해 URL 추출
        title = article.text  # 링크 텍스트를 통해 제목 추출
        
        # 추출된 URL과 제목을 리스트에 추가
        if url:  # href 속성이 존재할 경우에만 리스트에 추가
            url_list.append(url)
            title_list.append(title)
    
    # 수집된 데이터를 출력
    for title, url in zip(title_list, url_list):
        print(f"Title: {title}, URL: {url}")
    
    print(" ")
    print(f"URL 개수: {len(url_list)}")
    print(f"제목 개수: {len(title_list)}")
    
    # DataFrame으로 변환하여 저장
    df = pd.DataFrame({'URL': url_list, 'Title': title_list})
    print(df)
    df.to_csv(f"knowledge_links.csv", index=False)

except Exception as e:
    print(f"오류 발생: {e}")
finally:
    driver.quit()
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import json
from datetime import datetime

# CSV 파일 읽기
df = pd.read_csv('knowledge_links.csv')  

# JSON으로 변환할 빈 리스트
data_list = []
content_list = []

for _, row in df.iterrows():
    url = row['URL']
    title = row['Title']
    
    # URL에서 웹 페이지의 HTML 가져오기
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # 텍스트 내용 추출하기
        text_blocks = soup.find_all('p', {'class': 'txt'})  # 컨테이너 변경
        
        text = ' '.join([block.get_text(strip=True) for block in text_blocks])  # 텍스트 결합
        content_list.append(text)
        print(text)
        
        
        # 현재 날짜 가져오기
        now = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        
        # 데이터 구조 정의
        data = {
            'ID': _,  # 혹은 다른 고유 식별자
            'contents': {
                'title': title,
                'text': text
            },
            'link': url,
            'source': 'knowledge_blog',
            'date': now,
            'language': 'korean'
        }
        
        data_list.append(data)

# JSON 파일로 저장
with open('C:\\Users\\InterX_2team\\Desktop\\HB\\0401_Crawling\\knowledge_naver.json', 'w', encoding='utf-8') as f:
    json.dump(data_list, f, ensure_ascii=False, indent=4)
```
